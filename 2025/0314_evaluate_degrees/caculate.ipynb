```python
import pandas as pd
# 파일 로드
under = pd.read_csv('degrees_undergraduate.csv')
gradu = pd.read_csv('degrees_graduate.csv', encoding='cp949')
```

```python
# 성적이 F이거나 혹은 pass/fail 과목 제외
under = under[~ under['등급'].isin(['F', 'P'])]
gradu = gradu[(gradu['등급']!='F') & (gradu['이수구분']!='연구지도')]
# 재수강과목 기존 행 삭제
under = under[under['삭제구분']!='재수강']
display(under.head())
display(gradu.head(50))
```

```python
# 전체 평균 학점 계산
under_deg = (under['학점'] * under['평점']).sum() / under['학점'].sum()
gradu_deg = (gradu['학점'] * gradu['평점']).sum() / gradu['학점'].sum()
print("학부 평균 평점 : {0}".format(round(under_deg, 2)))
print("대학원 평균 평점 : {0}".format(round(gradu_deg, 2)))
```

```python
# 전공과목
under_major = under[under['이수구분'].isin(['전공필수', '전공선택'])]
gradu_major = gradu[gradu['이수구분'].isin(['전공선택', '전공'])]
display(under_major.head())
display(gradu_major.head())
```

```python
# 전공 평균 학점 계산
under_mdeg = (under_major['학점'] * under_major['평점']).sum() / under_major['학점'].sum()
gradu_mdeg = (gradu_major['학점'] * gradu_major['평점']).sum() / gradu_major['학점'].sum()
print("학부 평균 평점 : {0}".format(round(under_mdeg, 2)))
print("대학원 평균 평점 : {0}".format(round(gradu_mdeg, 2)))
```
